```python
# ── TEMPLATE DASHBOARD VISUALISASI STATIS ────────────────────────────
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, matplotlib.gridspec as gridspec
import seaborn as sns
from IPython.display import Markdown, display

# ── 1. LOAD & INSPECT DATASET Flights──────────────────────────────────────────
df = sns.load_dataset('tips')
print(df.shape, df.dtypes)
print(df.describe().round(2))

print(' ')
print(' ')

# ── 2. SETUP FIGURE LAYOUT ─────────────────────────────────────────────
fig = plt.figure(figsize=(16, 12))
fig.suptitle('Dashboard Analisis Dataset Tips',
fontsize=18, fontweight='bold', y=0.98)
# Grid 2 x 4 subplot
gs = gridspec.GridSpec(2, 2, figure=fig,
hspace=0.4, wspace=0.35)
ax1 = fig.add_subplot(gs[0, 0]) # kiri atas
ax2 = fig.add_subplot(gs[0, 1]) # kanan atas

ax3 = fig.add_subplot(gs[1, 0]) # kiri bawah
ax4 = fig.add_subplot(gs[1, 1]) # kanan bawah


# ── 34a. GRAFIK 1: Bar Chart ─────────────────────────────────────────────
avg_by_day = df.groupby('day')['total_bill'].mean()
avg_by_day.plot(kind='bar', ax=ax1, color='#028090',
edgecolor='white')
ax1.set_title('Rata-rata Total Tagihan per Hari')
ax1.set_xlabel('Hari'); ax1.set_ylabel('Total Bill (USD)')
ax1.tick_params(axis='x', rotation=0)
ax1.spines[['top','right']].set_visible(False)

# ── 34. GRAFIK 2: Histogram + KDE ───────────────────────────────────────
sns.histplot(data=df, x='total_bill', hue='time',
kde=True, palette='Set2', ax=ax2)
ax2.set_title('Distribusi Total Tagihan per Waktu Makan')
ax2.spines[['top','right']].set_visible(False)

# ── 34. GRAFIK 3: Boxplot ───────────────────────────────────────────────
sns.boxplot(data=df, x='day', y='total_bill',
hue='sex', palette='Set1', ax=ax3)
ax3.set_title('Distribusi Tagihan per Hari & Gender')
ax3.spines[['top','right']].set_visible(False)

# ── 34. GRAFIK 4: Scatter Plot ──────────────────────────────────────────
sns.scatterplot(data=df, x='total_bill', y='tip',
hue='time', size='size', sizes=(30,180),
palette='Set2', alpha=0.7, ax=ax4)
sns.regplot(data=df, x='total_bill', y='tip',
scatter=False, color='gray', ax=ax4)
ax4.set_title('Hubungan Tagihan & Tip (Ukuran = Jumlah Tamu)')
ax4.spines[['top','right']].set_visible(False)

# ── 5. EKSPOR ──────────────────────────────────────────────────────────
plt.savefig('dashboard_tips.png', dpi=150, bbox_inches='tight')
plt.show()
print('Dashboard tersimpan sebagai dashboard_tips.png')
print('')

markdown_table = """
| Pertanyaan | Narasi Bar Chart |
|------------|-----------------|
| **What?** | Grafik batang menunjukkan rata-rata total tagihan per hari. Nilai tertinggi terjadi pada hari Minggu sekitar **21,3**, sedangkan nilai terendah terjadi pada hari Jumat sekitar **17,0**. Hari Sabtu juga menunjukkan rata-rata yang tinggi yaitu sekitar **20,3**. |
| **So what?** | Data ini menunjukkan bahwa pengeluaran pelanggan meningkat pada akhir pekan, khususnya Sabtu dan Minggu. Hal tersebut penting karena dapat menjadi indikasi meningkatnya aktivitas pelanggan atau jumlah kunjungan saat akhir pekan dibandingkan hari kerja biasa. |
| **Now what?** | Pelaku usaha dapat memanfaatkan pola ini dengan meningkatkan persediaan, jumlah karyawan, atau membuat promosi khusus pada akhir pekan. Selain itu, analisis lanjutan dapat dilakukan untuk mengetahui faktor yang menyebabkan kenaikan tagihan pada hari Sabtu dan Minggu. |
"""

display(Markdown(markdown_table))


markdown_table = """
|   Pertanyaan  | Narasi Histogram |
|---------------|---------|
| **What?**     | Grafik histogram menunjukkan distribusi total tagihan berdasarkan waktu makan, yaitu **Lunch** dan **Dinner**. Sebagian besar total tagihan saat makan siang berada pada kisaran **10–20**, sedangkan makan malam memiliki distribusi yang lebih luas hingga mencapai sekitar **50**. Frekuensi tertinggi terlihat pada makan malam dengan total tagihan sekitar **15–20**. |
| **So what?**  | Data ini menunjukkan bahwa pelanggan cenderung mengeluarkan biaya lebih besar saat makan malam dibandingkan makan siang. Hal ini penting karena dapat mengindikasikan bahwa makan malam lebih sering digunakan untuk acara bersama keluarga, teman, atau pertemuan khusus sehingga total pembelanjaan meningkat. |
| **Now what?** | Restoran dapat memanfaatkan pola ini dengan menyediakan menu premium atau promo khusus saat makan malam untuk meningkatkan pendapatan. Selain itu, analisis lebih lanjut dapat dilakukan untuk mengetahui faktor lain yang memengaruhi tingginya total tagihan pada waktu dinner, seperti jumlah pelanggan per meja atau jenis menu yang dipesan. |
"""
display(Markdown(markdown_table))


markdown_table = """
| Pertanyaan | Narasi Boxplot |
|------------|---------|
| **What?** | Grafik boxplot menunjukkan distribusi total tagihan berdasarkan hari dan gender pelanggan. Secara umum, pelanggan laki-laki (**Male**) memiliki median total tagihan yang sedikit lebih tinggi dibandingkan perempuan (**Female**) pada hampir semua hari. Hari Minggu menunjukkan variasi tagihan paling besar dengan beberapa outlier mencapai sekitar **50**. |
| **So what?** | Data ini penting karena menunjukkan adanya perbedaan pola pengeluaran berdasarkan gender dan hari. Akhir pekan, terutama Sabtu dan Minggu, cenderung memiliki total tagihan yang lebih tinggi dan lebih bervariasi, yang dapat mengindikasikan peningkatan aktivitas pelanggan pada hari tersebut. |
| **Now what?** | Restoran dapat memanfaatkan informasi ini untuk menyusun strategi layanan atau promosi khusus pada akhir pekan. Analisis lanjutan juga dapat dilakukan untuk mengetahui apakah faktor lain, seperti jumlah pengunjung per meja atau jenis pesanan, memengaruhi tingginya total tagihan pada pelanggan tertentu. |
"""

display(Markdown(markdown_table))

markdown_table = """
| Pertanyaan | Narasi Scatter Plot|
|------------|---------|
| **What?** | Grafik scatter plot menunjukkan hubungan antara **total tagihan (total_bill)** dan **tip** dengan ukuran titik mewakili jumlah tamu (**size**) serta warna berdasarkan waktu makan (**Lunch** dan **Dinner**). Terlihat adanya tren positif, di mana semakin besar total tagihan, maka nilai tip juga cenderung meningkat. Sebagian besar data terkonsentrasi pada total tagihan sekitar **10–30** dengan tip sekitar **2–5**. |
| **So what?** | Grafik ini penting karena menunjukkan bahwa pelanggan yang mengeluarkan tagihan lebih besar cenderung memberikan tip lebih tinggi. Selain itu, makan malam (**Dinner**) tampak memiliki lebih banyak titik dengan total tagihan dan tip tinggi dibandingkan makan siang (**Lunch**), yang dapat menunjukkan potensi pendapatan lebih besar pada waktu dinner. |
| **Now what?** | Restoran dapat memanfaatkan informasi ini untuk meningkatkan kualitas layanan pada pelanggan dengan jumlah tamu lebih besar atau pada waktu dinner karena berpotensi menghasilkan tip lebih tinggi. Analisis lanjutan juga dapat dilakukan untuk mengetahui apakah ukuran kelompok pelanggan memiliki pengaruh signifikan terhadap besarnya tip yang diberikan. |
"""

display(Markdown(markdown_table))
```
