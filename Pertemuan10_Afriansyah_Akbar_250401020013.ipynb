```python
import pandas as pd

# 1
print('Muat dan Eksplorasi Data')
print(' ')
url = "https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv"
df = pd.read_csv(url)
print("=== DIMENSI DATA ===")
print(f"Dimensi dataset: {df.shape}\n")
print("=== PROPORSI KELAS TARGET (CHURN) ===")
print(df["Churn"].value_counts(normalize=True))
print("-" * 50)

# 2
print(' ')
print('Preprocessing')
print(' ')
from sklearn.model_selection import train_test_split

# Pembersihan Data: Mengubah 'TotalCharges' yang kosong (berisi spasi) menjadi NaN dan menghapusnya
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df = df.dropna()
# Menghapus kolom 'customerID' karena tidak memiliki nilai prediktif
X = df.drop(columns=['customerID', 'Churn'])
# Mengubah target 'Churn' menjadi numerik (Yes = 1, No = 0)
y = df['Churn'].map({'Yes': 1, 'No': 0})
# Encoding fitur kategorikal menggunakan One-Hot Encoding (pd.get_dummies)
X = pd.get_dummies(X, drop_first=True)
# Membagi data menjadi data latih (80%) dan data uji (20%) secara stratified
X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("=== PREPROCESSING SELESAI ===")
print(f"Jumlah data latihan (X_tr): {X_tr.shape[0]} baris")
print(f"Jumlah data pengujian (X_te): {X_te.shape[0]} baris\n")
print("-" * 50)

# 3
print(' ')
print('Latih Model')
print(' ')
from sklearn.ensemble import RandomForestClassifier
# Membangun model RandomForestClassifier dengan penanganan class_weight="balanced"
rf = RandomForestClassifier(n_estimators=300, class_weight="balanced", random_state=42)
# Melatih model menggunakan data latihan
rf.fit(X_tr, y_tr)
print("=== MODEL BERHASIL DILATIH ===\n")
print("-" * 50)

# 4 dan 5
print(' ')
print('Evaluasi dan Prediksi Probabilitas serta Kesimpulkan')
print(' ')
from sklearn.metrics import classification_report, roc_auc_score
# Menghitung prediksi label (0 atau 1)
y_pred = rf.predict(X_te)
# Menghitung probabilitas untuk kelas positif (Churn = 1)
proba = rf.predict_proba(X_te)[:, 1]
print("=== EVALUASI MODEL ===")
print("Classification Report:")
print(classification_report(y_te, y_pred))
# Menghitung nilai ROC-AUC
roc_auc = roc_auc_score(y_te, proba)
print(f"ROC-AUC Score: {roc_auc:.4f}\n")
print("-" * 50)
```

=== KESIMPULAN UTAMA MODEL ===

1. Dataset Telco Customer Churn berhasil diproses melalui tahap preprocessing, seperti pembersihan data, encoding fitur kategorikal, dan pembagian data latih serta data uji sehingga siap digunakan untuk pemodelan.
2. Model Random Forest berhasil dilatih menggunakan data latih dan mampu melakukan klasifikasi pelanggan yang berpotensi melakukan churn berdasarkan fitur-fitur yang tersedia.
3. Kinerja model dievaluasi menggunakan Classification Report dan ROC-AUC Score, sehingga dapat diketahui tingkat akurasi serta kemampuan model dalam membedakan pelanggan yang akan churn dan yang tidak.
