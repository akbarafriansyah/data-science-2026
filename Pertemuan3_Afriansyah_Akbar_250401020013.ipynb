```python
#Praktikum Pertemuan 3
#2
import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize

#3
URL = "https://drive.google.com/uc?id=1LfQWProB0VjWN5q8bKuRIgn-stULfIRo"

#4
df = pd.read_csv(URL)

print(df.info())
print(' ')
print(df.describe())

print(' ')
print('Shape awal:', df.shape)
print(df.isnull().sum())

#5
print(' ')
df.drop_duplicates(inplace=True)
print('Setelah hapus duplikat:', df.shape)

#6
print(' ')
df['kota'] = df['kota'].str.strip().str.title()
df['kondisi'] = df['kondisi'].str.strip().str.lower()

#7
print(' ')
df['luas_m2'] = df['luas_m2'].fillna(df['luas_m2'].median())
df['harga_juta'] = df['harga_juta'].fillna(df['harga_juta'].median())
df['kamar'] = df['kamar'].fillna(df['kamar'].mode()[0])

#8
print(' ')
for col in ['harga_juta', 'luas_m2', 'tahun_bangun']:Q1, Q3 = df[col].quantile([0.25, 0.75])
IQR = Q3 - Q1
df[col] = df[col].clip(Q1-1.5*IQR, Q3+1.5*IQR)

#9
print(' ')
assert df.isnull().sum().sum() == 0, 'Masih ada missing!'
assert df.duplicated().sum() == 0, 'Masih ada duplikat!'
print('Shape akhir:', df.shape)

#10
print(' ')
df.to_csv('housing_clean.csv', index=False)
print('Dataset bersih tersimpan!')

#11
print(' ')
import requests

# Mengakses API JSONPlaceholder
url_api = "https://jsonplaceholder.typicode.com/posts"
response = requests.get(url_api)

# Mengubah response JSON menjadi DataFrame
data_json = response.json()
df_api = pd.DataFrame(data_json)

# Menampilkan informasi DataFrame
print(' ')
print(df_api.head())
print(' ')
print(df_api.info())

# Menyimpan DataFrame ke CSV
df_api.to_csv('jsonplaceholder_posts.csv', index=False)
print('Data API berhasil disimpan!')
```
