```python
#Praktikum Pertemuan 2
import pandas as pd
import numpy as np

#Muat Dataset Titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)

#Eksplorasi Awal
print(df.shape, df.columns.tolist())
print(df.isnull().sum())

# 3 Filtering — Penumpang Wanita Kelas 1
wk1 = df[(df['Sex']=='female') &
(df['Pclass']==1)]

print(wk1['Survived'].mean(), wk1['Age'].mean())

# 4 Groupby — Tingkat Keselamatan per Kelas
hasil = df.groupby('Pclass').agg(
total=('PassengerId','count'),
selamat=('Survived','mean'),
).round(3)
print(hasil)
```

```python
#Q1
print("Jumlah Baris dan Kolom     : ", df.shape)
print("Kolom yang tersedia adalah : ", df.columns.tolist())
```

Dataset Titanic memiliki **891 baris dan 12 kolom**.
Artinya terdapat 891 data penumpang dengan 12 variabel yang mendeskripsikan setiap penumpang.

```python
#Q2
missing = df.isnull().sum()
percent = (missing / len(df)) * 100

pd.DataFrame({
    'Jumlah Missing': missing,
    'Persentase (%)': percent
})
```

Ditemukan bahwa kolom **Age, Cabin, dan Embarked** memiliki nilai kosong.
Kolom Cabin memiliki jumlah missing paling banyak.

1.   Cabin (77.10%)
2.   Age (19.86%)
3.   Embarkked (0.22%)



```python
#Q3
print("Persentase penumpang yang selamat sebanyak", df['Survived'].mean() * 100,"%")
```

Berdasarkan hal tersebut didapat sekitar **38,38%** penumpang berhasil selamat dari kecelakaan Titanic.

```python
#Q4
female_class1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]

jumlah = female_class1.shape[0]
rata_usia = female_class1['Age'].mean()

print("Jumlah Penumpang Wanita dari kelas 1 berjumlah",jumlah,"orang, dan rata-rata usia yang selamat adalah sebesar", rata_usia,"%")
```

Hasil yang didapat adalah Jumlah Penumpang wanita kelas 1 yang selamat adalah **94 orang** dan rata - rata usia adalah sekitar 34.9 Tahun atau Rata-rata usia apat dibulatkan menjadi **35 tahun.**

```python
#Q5
survival_by_class = df.groupby('Pclass')['Survived'].mean()*100
print("Kelas Penumpang dengan tingkat keselamat tertinggi adalah kelas (dalam %):")
print("")
survival_by_class
```

Kelas 1 memiliki tingkat keselamatan tertinggi dibandingkan kelas lainnya yaitu dengan nilai rata - rata **0.63%**
didapat juga kesimpulan bahwa semakin tinggi kelas, semakin besar kemungkinan untuk selamat.

```python
#Q6
print("Variabel yang saya ambil merupakan tingkat keselamatan dalam kategori jenis kelamin:")
print("")
survival_by_gender = df.groupby('Sex')['Survived'].mean()*100
survival_by_gender
```

Interpretasi saya berdasarkan analisis ini adalah:
Wanita jauh lebih banyak selamat dibanding pria,
Ini menunjukkan adanya kebijakan "women and children first" saat evakuasi
Faktor sosial sangat berpengaruh, bukan hanya kelas ekonomi.
